```python

import numpy as np
import time
from tqdm import tqdm
```

```python
def sig_round_off(number,digits):
    return np.round(number+5*10**(-1*(digits+1)),decimals=digits-1)
```

```python
def Cal_LU_pivot(D,g,d):
    A= np.array((D),dtype=float)
    f= np.array((g),dtype=float)
    n = len(f)
    time_for_partial_pivoting = 0
    time_for_factor_calculation = 0
    time_for_addition_operation = 0
    for i in tqdm(range(0,n-1)):     # Loop through the columns of the matrix
        for k in range(i+1,n):
            start_partial_pivoting = time.time()
            if np.abs(A[k,i])>np.abs(A[i,i]):
                A[[i,k]]=A[[k,i]]             # Swaps ith and kth rows to each other
                f[[i,k]]=f[[k,i]]
            time_for_partial_pivoting += (time.time() - start_partial_pivoting)  
        for j in range(i+1,n):# Loop through rows below diagonal for each column
            start_factor = time.time()
            m = A[j,i]/A[i,i]
            m = sig_round_off(m,d)
            time_for_factor_calculation += (time.time() - start_factor)
            start_sum = time.time()
            A[j,:] = sig_round_off(A[j,:] - m*A[i,:],d)
            f[j] = sig_round_off(f[j] - m*f[i],d)
            time_for_addition_operation += (time.time() - start_sum)
    print(f'Time taken for partial pivoting: {time_for_partial_pivoting}')
    print(f'Time taken for multiplication: {time_for_factor_calculation}')
    print(f'Time taken for addition: {time_for_addition_operation}')
    return A,f
```

```python
def Back_Subs(A,f,d):
    n = f.size
    x = np.zeros(n)# Initialize the solution vector, x, to zero
    start_divison = time.time()
    x[n-1] = f[n-1]/A[n-1,n-1]    # Solve for last entry first
    print(f'Time taken for divison:{time.time() - start_divison}')
    time_addition_prod = 0
    for i in tqdm(range(n-2,-1,-1)):# Loop from the end to the beginning
        start_addition_prod = time.time()
        sum_ = 0
        for j in range(i+1,n):        # For known x values, sum and move to rhs
            sum_ = sig_round_off(sum_ + A[i,j]*x[j],d)
        time_addition_prod +=  (time.time() - start_addition_prod)
        x[i] = sig_round_off((f[i] - sum_)/A[i,i],d)
    print(f'Time taken for addition and multiplication: {time_addition_prod}')
    return x
```

```python
def gauss_elimination_with_pivot(A,f,D):
    start_forward_elimination = time.time()
    B,g = Cal_LU_pivot(A,f,D)
    print(f'Time Taken for forward Elimination:{time.time() - start_forward_elimination}')
    start_back_substitution = time.time()
    X = Back_Subs(B,g,D)
    print(f'Time Taken for Backward Substituion:{time.time() - start_back_substitution}')
    return A,f,X
```

```python
def lu(A):
    
    #Get the number of rows
    n = A.shape[0]
    
    U = A.copy()
    L = np.eye(n, dtype=np.double)
    
    #Loop over rows
    for i in tqdm(range(n)):
            
        #Eliminate entries below i with row operations 
        #on U and reverse the row operations to 
        #manipulate L
        factor = U[i+1:, i] / U[i, i]
        L[i+1:, i] = factor
        U[i+1:] -= factor[:, np.newaxis] * U[i]
        
    return L, U
```

```python
def forward_elimination(L, b):
    
    #Get number of rows
    n = L.shape[0]
    
    #Allocating space for the solution vector
    y = np.zeros_like(b, dtype=np.double);
    
    #Here we perform the forward-substitution.  
    #Initializing  with the first row.
    y[0] = b[0] / L[0, 0]
    
    #Looping over rows in reverse (from the bottom  up),
    #starting with the second to last row, because  the 
    #last row solve was completed in the last step.
    for i in tqdm(range(1, n)):
        y[i] = (b[i] - np.dot(L[i,:i], y[:i])) / L[i,i]
        
    return y
```

```python
def back_substitution(U, y):
    
    #Number of rows
    n = U.shape[0]
    
    #Allocating space for the solution vector
    x = np.zeros_like(y, dtype=np.double);

    #Here we perform the back-substitution.  
    #Initializing with the last row.
    x[-1] = y[-1] / U[-1, -1]
    
    #Looping over rows in reverse (from the bottom up), 
    #starting with the second to last row, because the 
    #last row solve was completed in the last step.
    for i in tqdm(range(n-2, -1, -1)):
        x[i] = (y[i] - np.dot(U[i,i:], x[i:])) / U[i,i]
        
    return x
```

```python
def lu_solve(A, b):
    start_lu = time.time()
    L, U = lu(A)
    print(f"Time taken for Decomposition  {time.time() - start_lu}")
    start_forward_elimination = time.time()
    y = forward_elimination(L, b)
    print (f"solve time for forward  {time.time()-start_forward_elimination}")
    solve_backward_substitution = time.time()
    z = back_substitution(U, y)
    print (f"solve time for backword substituion  {time.time() - solve_backward_substitution}")
    return z
```
