From the plot we can see that the price increses w.r.t the increase in Engine-size. Hence a positive correlation is seen here

```python
df_corr_2 = df[["highway-mpg", "price"]]
df_corr_2.head()
```

```python
df_corr_2.plot(kind='scatter', x='highway-mpg', y='price', figsize=(10, 6), color='red')

plt.title('Highway-mpg Vs Price')
plt.xlabel('Highway-mpg')
plt.ylabel('Price')

plt.show()
```

From the plot we can see that the price decreases with incresae in highway mpg. Hence a negative correlation exists between these two variables

```python
df_corr_3 = df[["peak-rpm", "price"]]
df_corr_3.head()
```

```python
df_corr_3.plot(kind='scatter', x='peak-rpm', y='price', figsize=(10, 6), color='darkblue')

plt.title('peak-rpm Vs Price')
plt.xlabel('peak-rpm')
plt.ylabel('Price')

plt.show()
```

There is no positive or negative trend seen in the plot. Hence no correlation exists between the two variables.

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12))
sns.heatmap(df[['engine-size', 'price']].corr(), ax=ax1, annot=True, vmin=-1, vmax=1)
sns.heatmap(df[["highway-mpg", "price"]].corr(), ax=ax2, annot=True, vmin=-1, vmax=1)
sns.heatmap(df[["peak-rpm", "price"]].corr(), ax=ax3, annot=True, vmin=-1, vmax=1)

plt.show()
```

### Observations

1. Among engine-size, highway-mpg and peak-rpm, engine-size and highway-mpg are the potential predictor variable of price as they are highly correlated to price.
2. Pre-attentive attributes used are:
    1. Spatial Positioning: in scatter plot to determine weather there is any correlation
    2. Color: in heatmap to determine how much features are correlated to price
3. Gestalt principles used are:
    1. Proximity: Through the position of near points in scatter plot, we can say in which direction points are going
    2. Similarity:  Through the colour we can tell that less correlated features are of pink shade, but correlated features are more white or brown

### Observations

1. Relationship between:
    1. body-style and price: `convertible` and `hardtop` style cars are on average more expensive than `hatchback`, `sedan` and `wagon`, these three are similarly priced with `hatchback` being the cheapest.
    2. engine-location and price: other than outliers in `front`, cars with `rear` engine are always more expensive `front` engine
    3. drive-wheels and price: `rwd` are generally more expensive the `fwd` and `4wd`, both of which are similarly priced but `fwd` has more range than `4wd`
2. Pre-attentive attributes used are:
    1. Form: size of the box determines the range of price from 1st quartile to 3rd quartile. Dots denotes the outlier.
    2. Spatial Positioning: position of box denotes the price in respective quartile, position of dots shows how big of outlier they are and their price.
3. Gestalt principles used are:
    1. Proximity: Through the position of position of box, whisker and dots we can determine their price and where they stand compared to other categories
    2. Figure and Ground: Box plots are the figures and grid lines are the background helping in comparing the categories.

### Question 5
<h4> Use a single heat map to visualize the relationship between Body Style and Drive-wheel vs Price.</h4>
<h4>Perform the data preprocessing required.<br>

<p>The heatmap should plot  the target variable (price) proportional to colour with respect to the variables 'drive-wheel' and 'body-style' in the vertical and horizontal axis respectively. This allows us to visualize how the price is related to 'drive-wheel' and 'body-style'.</p>

The plot should have labels for x axis and y axis<br>
Write down your observations in a markdown cell below the visual</h4>

1.Mention the pre-attentive attributes used.(atleast 2)<br>
2.Mention the gestalt principles used.(atleast 2)<br>
   


```python
sns.set(font_scale=1.4)

body_style_df = df.pivot_table(values='price', index='body-style', aggfunc=np.mean).sort_values(by='price', axis=0, ascending=False)
drive_wheel_df = df.pivot_table(values='price', index='drive-wheels', aggfunc=np.mean).sort_values(by='price', axis=0, ascending=False)
fig, axis = plt.subplots(1, 2, figsize=(15, 10))
sns.heatmap(body_style_df, vmin=23000, vmax=9000, annot=True, ax=axis[0], annot_kws={"fontsize": 20})
sns.heatmap(drive_wheel_df, vmin=23000, vmax=9000, annot=True, ax=axis[1], annot_kws={"fontsize": 20})

plt.show()
```

```python
df_heatmap = df.pivot_table(values='price', index='body-style', columns='drive-wheels', aggfunc=np.mean)
fig, axis = plt.subplots(figsize=(15, 10))
sns.heatmap(df_heatmap, annot=True, ax=axis, annot_kws={"fontsize": 20}, cbar_kws={'label': 'price'})

plt.show()
```

### Observations

1. Through above visuals we can conclude that:
    1. `convertible` & `hardtop` in body-style and `rwd` in drive-wheels leads to higher price so combination of these 2 will result in most expensive cars(we can confirm this from second visuals)
    2. `4wd` & `fwd` in drive-wheels leads to cheaper priced cars
2. Pre-attentive attributes used are:
    1. Color: denotes how much the feature category contributes to price increase
    2. Spatial Positioning: visuals have been sorted to how higher priced categories over lower priced categories.
3. Gestalt principles used are:
    1. Similarity:  Through the colour we can tell that lesser priced categories are of blue shade, but higher priced categories are more brown
    2. Similarity: Thorough similar colours between categories of body-style and drive-wheels we can tell which categories are similarly priced.
